```python
import pandas as pd
```

```python
# import csvs and display first few elements
train_data = pd.read_csv("train.csv")
train_data.head()
```

```python
# Pre-processing - lowercase, punctuation, stemming, removing stop words
train_data['review'] = train_data.review.map(lambda x: x.lower())
train_data['review'] = train_data.review.str.replace('[^\w\s]','')
train_data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train_data['review'], train_data['sentiment'], test_size=0.1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
text_clf = Pipeline([
    ('vectorizer', CountVectorizer(stop_words = "english")),
    ('tfidf', TfidfTransformer()),
    ('clf', MultinomialNB()),
])
text_clf.fit(train_data['review'], train_data['sentiment'])
```

```python
import pickle
classifier = "amazon_naive_bayes.pkl"

with open(classifier, 'wb')as file:
    pickle.dump(text_clf, file)
```

```python
test_data = pd.read_csv("test.csv")
test_data.head()

text_clf.predict(test_data['review'])
```
